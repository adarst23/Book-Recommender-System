# Collaborative Filtering Recommender System

We will select only that users that have rated more than 250 books <br>
& only that books that have atleast 50 ratings

```python
ratings_books.groupby('User-ID').count()['Book-Rating']
```

```python
temp[temp]
```

```python
temp = ratings_books.groupby('User-ID').count()['Book-Rating'] > 200

required_user = temp[temp].index   # These are the users that have more given more than 200 ratings
```

```python
ratings_books[ratings_books['User-ID'].isin(required_user)]   #It is book table that is rated by above 811 raters
```

```python
filtered_rating = ratings_books[ratings_books['User-ID'].isin(required_user)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
```

```python
y
```

```python
famous_books = y[y].index   # Books that have more than 50 ratings & rated by raters that are among 811 raters(200 rates)
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]  #Books that have more than 50 ratings
```

```python
final_ratings
```

```python
final_ratings.duplicated().sum()
```

```python
df = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
df.fillna(0,inplace=True)
```

```python
df.head()    #It is a data in which each user have rated more than 200 books & each books have more than 50 ratings
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(df)
```

```python
similarity.shape
```

```python
similarity
```

```python
def recommend(book_name):
    #index fetch
    index = np.where(df.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity[index])), key=lambda x: x[1], reverse=True)[1:6]
    
    for i in similar_items:
        print(df.index[i[0]])
    
```

```python
recommend('1984')
```

```python
books.loc[books['Book-Title'] == '1984'].index[0]
```

```python
recommend('The Da Vinci Code')
```
